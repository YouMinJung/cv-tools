```python
import os

import cv2
```

```python
VIDEO_PATH = '/mnt/team-experiments/robotech_test/robotech_test3.mp4'
OUT_PATH = '/mnt/hdd1/robotech_test'
video = cv2.VideoCapture(VIDEO_PATH)

length = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(video.get(cv2.CAP_PROP_FPS))

print(f'Total Frames: {length}')
print(f'Width: {width}, Height: {height}')
print(f'FPS: {fps}')

outpath = VIDEO_PATH[:-4]
if OUT_PATH:
    outpath = os.path.join(OUT_PATH, os.path.basename(VIDEO_PATH)[:-4])

try:
    if not os.path.exists(outpath):
        os.makedirs(outpath, exist_ok=True)
except OSError:
    print(f'Error: Creating directory. {outpath}')


success, image = video.read()
count = 0
fps = 1
while success:
    success, image = video.read()

    if (int(video.get(cv2.CAP_PROP_POS_FRAMES)) % fps == 0):
        fname = f'{outpath}/{count:06d}.jpg'
        cv2.imwrite(fname, image)
        print(f'Read a new frame: {fname}')
        count += 1
print('Finish! convert video to frame')

video.release()
```
